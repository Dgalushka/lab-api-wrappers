### function that will add 10 unique tracks id's of each artist in the playlist

```python
def tentracks_of_each_artist(playlist_id): 
    playlist = sp.user_playlist_tracks(playlist_id = playlist_id)
    
    artists_set = {playlist["items"][i]['track']['album']['artists'][0]['name'] for i in range(playlist["total"])}
    artists_list = list(artists_set)
    
    track_ids = []
    for artist in artists_list:
        result = sp.search(q=f"artist:{artist}", limit=10, type='track')
        for track in result["tracks"]["items"]:
            track_ids.append(track["uri"])
    #print(len(list(set(track_ids))))
    return list(set(track_ids)) #set is used to make sure there are no dublicates
    
    
```

```python
tentracks_of_each_artist("37i9dQZEVXbLiRSasKsNU9") # viral top-50 global playlist
```

```python
pd.json_normalize(sp.audio_features(track_ids[100:200]))
```

```python
start = 0
df = pd.DataFrame()

for stop in range(0, len(track_ids)+100, 100):
    if start != stop:
        print(start, stop)
        new_df = pd.json_normalize(sp.audio_features(track_ids[start:stop]))
        df = pd.concat([new_df, df])
        start = stop

df.reset_index()
```

```python
start = 0
for stop in range(0, len(tentracks_of_each_artist("37i9dQZEVXbLiRSasKsNU9"))+100, 100):
    if start != stop:
        print(start, stop)
        new_df = pd.json_normalize(sp.audio_features(tentracks_of_each_artist("37i9dQZEVXbLiRSasKsNU9")[start:stop]))
        df = pd.concat([new_df, df])
        start = stop
    
```
